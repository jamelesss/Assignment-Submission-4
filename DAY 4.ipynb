```python
class School:
  def __init__(self, name,design, weight,):
    self.name = name 
    self.design = design
    self.weight = weight
    
p1 = School("Tom Snow", "Student", 32)
p2 = School("John Smith", "Teacher", 48)

print(p1.name, p1.design, p1.weight)
print(p2.name, p2.design, p2.weight)
```
